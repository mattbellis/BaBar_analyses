```python
mask = (df_col['cut_2']==True) & (df_col['cut_3']==True)
#mask = mask & (df_col['cut_4']==True)

x_col = df_col[mask]['BpostFitMes']
y_col = df_col[mask]['BpostFitDeltaE']

plt.plot(x_col,y_col,'.', alpha=0.2)
```

```python
mask_col = (df_col['cut_2']==True) & (df_col['cut_3']==True)
#mask_col = mask_col & (df_col['cut_4']==True)

mask_sp = (df_sp['cut_2']==True) & (df_sp['cut_3']==True)
mask_sp = mask_sp & (df_sp['spmode']!='-999')
#mask_sp = mask_sp & (df_sp['cut_4']==True)


#var = 'BcosthCM'
#lo,hi = -1,1

var = 'BpostFitMes'
lo,hi = 5.2, 5.3

x_col = df_col[mask_col][var]
x_sp = df_sp[mask_sp][var]

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.hist(x_col,bins=100, range=(lo,hi))
plt.xlabel(var);

plt.subplot(1,3,2)
plt.hist(x_sp,bins=100, range=(lo,hi))
plt.xlabel(var);

plt.subplot(1,3,3)
plt.hist(x_sp,bins=100, range=(lo,hi), density=True, alpha=0.5)
plt.hist(x_col,bins=100, range=(lo,hi), density=True, alpha=0.5)
plt.xlabel(var);
```
